```python
#Q4 Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
```

```python
import matplotlib.pyplot as plt
file_path = "C:/Users/vennela/Downloads/Dataset.txt"
x_label = 'Values on X-axis'
y_label = 'Values on Y-Axis'
title = 'Line Plot of the Values'
with open(file_path, 'r') as file:
    lines = file.readlines()
    data = [list(map(float, line.strip().split())) for line in lines]
x_values = [item[0] for item in data]
y_values = [item[1] for item in data]
plt.plot(x_values, y_values, label='Line Plot',color='g')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.legend()
plt.show()
```

```python
#Q7 Create plot using subplots(), scatter, and some other methods on the Axes object:
```

```python
import matplotlib.pyplot as plt
import numpy as np
x, y = np.random.rand(50), np.random.rand(50)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Scatter plot
axs[0, 0].scatter(x, y, color='blue', label='Scatter Plot')

# Line plot
axs[0, 1].plot(x, y, color='orange', linestyle='--', marker='o', label='Line Plot')

# Bar chart
bar_data = {'Category 1': 20, 'Category 2': 35, 'Category 3': 15}
axs[1, 0].bar(bar_data.keys(), bar_data.values(), color='red', label='Bar Chart',edgecolor='black',width=0.5)

# Histogram
axs[1, 1].hist(np.random.randn(1000), bins=30, color='yellow', label='Histogram',edgecolor='black')
plt.show()
```

```python
#Q8 Generate random data for x and y and create a scatter plot using Matplotlib.
```

```python

np.random.seed(42)
num_points = 100
x_data = np.random.rand(num_points)
y_data = 2 * x_data + 1 + 0.1 * np.random.randn(num_points)  # Linear relationship with some noise
plt.scatter(x_data, y_data, label='Data')
plt.xlabel('Values on X-axis')
plt.ylabel('Values on Y-axis')
plt.title('Scatter Plot of Random Data')
plt.legend()
plt.show()
```

```python
#Q11 Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
```

```python
import matplotlib.pyplot as plt
categories = ['Rent', 'Utilities', 'Groceries', 'Entertainment', 'Transportation']
expenses = [1200, 400, 500, 750, 150]
myexplode=[0.02,0.02,0.02,0.01,0.01]
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['red', 'lightgreen', 'lightcoral', 'gold', 'lightpink'],explode=myexplode)
plt.title('Expense Distribution in Budget')
plt.show()
```
